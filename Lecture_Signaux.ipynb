# TP d'aéroacoustique : interaction écoulement/barreau

Ce notebook permet de relire un fichier contenant un signal de pression, de cacluler sa densité spectrale de puissance et de l'afficher. 

### Lecture du fichier

```python
#%matplotlib inline
%matplotlib notebook
import pylab as plt
import numpy as np

from scipy import signal
Gmems = np.sqrt(2)/420428
```

```python
fichier = "Interaction_ecoul_cyl_v_20.npy"
sig = Gmems * np.load(fichier)
Nt = np.shape(sig)[0]
Fs = 5e4
```

### Calcul de la dsp

```python
f2, sw = signal.welch(sig[:], Fs, nperseg=2*512) 
```

```python
plt.figure()
plt.plot(f2, sw, lw=2)
plt.grid()
plt.xlim([0,2000])
plt.xlabel(r"$f$ [Hz]")
plt.ylabel(r"$dsp$ ")
```

```python
def passe_haut(f,fc):
    I=np.where(f>fc)
    w = np.zeros_like(f)
    w[I]=1
    return w
w = passe_haut(f2,250)
plt.figure()
plt.plot(f2,w)
plt.grid()
plt.xlim([0,2000])
plt.xlabel(r"$f$ [Hz]")
plt.ylabel(r"$W$ ")
```
